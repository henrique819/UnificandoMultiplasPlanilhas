### UNIFICANDO DIVERSOS ARQUIVOS EXCEL COM DIVERSAS ABAS EM UM ÚNICO ARQUIVO EXCEL

```python
#Importar módulos
import pandas as pd
import glob
  
# Definindo o ditretório onde estão os arquivos em .XLSX
path = r'C:\Users\luis Shikasho\Documents\excels\teste'
  
# Importando todos os arquivos no formato excel, que estão no diretório C:\Users\luis Shikasho\Documents\excels\teste
filenames = glob.glob(path + "\*.xlsx")
print('Arquivos Importados:', filenames)
  
# Inicializando um DataFrame vazio
finalexcelsheet = pd.DataFrame()
  
# Iterar em cada arquivo excel "um por um"
# Dentro do diretório : C:\Users\luis Shikasho\Documents\excels\teste
for file in filenames:
  
   # Combinando várias planilhas do Excel
   # Num único DataFrame
    df = pd.concat(pd.read_excel(file, sheet_name=None), ignore_index=True, sort=False)
  
    # Anexando os arquivos excel um por um
    finalexcelsheet = finalexcelsheet.append(df, ignore_index=True)

# Imprimir resultado
print('Final Sheet:')
display(finalexcelsheet)

#Salvando em um único arquivo excel chamado "Final.xlsx"
finalexcelsheet.to_excel(r'C:\Users\luis Shikasho\Documents\excels\teste\Final.xlsx', index=False)
```
